```python
def plot_rolling_functions(series, window=2):
    #series.rolling.median(series, window_size).plot(label="median")
    rolling_mean = series.rolling(window=window).mean().plot(label='mean')
    rolling_median = series.rolling(window=window).median().plot(label='median')
    std = series.rolling(window=window).std().plot(label='std')
    #skew = series.rolling(window=window).skew().plot(label='skew')
    #kurt = series.rolling(window=window).kurt().plot(label='kurt')
    #mini = series.rolling(window=window).min().plot(label='min')
    #maxi = series.rolling(window=window).max().plot(label='max')
   
    pyplot.title("Various rolling window functions, window size %s" % (window))
    print(rolling_mean)
    pyplot.legend()
    pyplot.show()
```
